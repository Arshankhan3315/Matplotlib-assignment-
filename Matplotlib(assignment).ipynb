# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

# Answer-1- Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in data analysis and data visualization. It provides a wide range of functions and methods to generate various types of plots, charts, and graphs, making it a fundamental tool for data scientists, engineers, researchers, and analysts.

# Line Plot: A line plot is a graph that displays data points connected by straight lines, typically used to represent trends or continuous data.

# Scatter Plot: A scatter plot displays individual data points as markers, useful for showing the relationship between two variables.

# Bar Plot: A bar plot displays data using rectangular bars with lengths proportional to the values they represent, often used to compare values across categories.

# Histogram: A histogram displays the distribution of a dataset by dividing it into bins and displaying the frequency or count of data points in each bin.

# Pie Chart: A pie chart displays data as slices of a circular pie, with each slice representing a portion of the whole data set. It's often used to show the composition or proportions of different categories within a dataset.

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

# Answer-3- The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange plots in a grid-like layout, specifying the number of rows, number of columns, and the position of each plot within the grid. This function is helpful for comparing multiple plots or displaying related data in a structured and organized manner.

```python
x = np.array([0, 1, 2, 3, 4, 5])
```

```python
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')
```

```python
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')
```

```python
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')
```

```python
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')
plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Answer-4-

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(10, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Profit Comparison by Company (Bar Plot)')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Profit Comparison by Company (Horizontal Bar Plot)')
plt.tight_layout()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# answer-5-A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution and statistical summary of a dataset. It is useful for visualizing the central tendency, spread, and skewness of the data, as well as identifying outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.grid(True)
plt.tight_layout()
```
